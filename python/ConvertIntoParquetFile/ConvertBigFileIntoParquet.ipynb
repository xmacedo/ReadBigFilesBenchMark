```python
import pandas as pd  

# Configurações
TXT_FILE = "big_file.txt"
PARQUET_FILE = "big_file.parquet"

# Carregar o arquivo TXT
df = pd.read_csv(TXT_FILE, sep=";", header=None, names=["Property", "Min", "Mean", "Max"])

# Salvar como Parquet
df.to_parquet(PARQUET_FILE, engine="pyarrow", index=False)

print(f"Arquivo convertido com sucesso: {PARQUET_FILE}")
```
