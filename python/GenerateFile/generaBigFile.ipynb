```python
import random

# Configs
NUM_LINES = 1_000_000_000  # 1 billion of lines
FILE_NAME = "big_file.txt"
NUM_PROPERTIES = 10  # Number of Properties per line

def generate_property():
    """Generates Min/Mean/Max values ​​for a property."""
    min_value = round(random.uniform(50_000, 100_000), 1)
    max_value = round(random.uniform(200_000, 400_000), 1)
    mean_value = round((min_value + max_value) / 2, 1)
    return f"{min_value}/{mean_value}/{max_value}"

def generate_line(line_number):
    """Generates a line in the expected format."""
    properties = [f"PROPERTY_{i}={generate_property()}" for i in range(NUM_PROPERTIES)]
    return "{" + ", ".join(properties) + "}\n"

# Writing to the file
with open(FILE_NAME, "w", buffering=10_000_000) as file:
    for i in range(NUM_LINES):
        file.write(generate_line(i))

print(f"File '{FILE_NAME}' successfully generated!")
```
